```python
import pandas as pd
import numpy as np

station = "mrlb"
data = pd.read_csv(f"../data/{station}/metars.csv", index_col=0)
data.head()
```

```python
df = data[["date", "visibility_m", "weather_1", "weather_2", "weather_3"]]
df = df.query("visibility_m < 8000.0")
df.head()
```

```python
df['precipitation'] = 0
df.loc[df['weather_1'].astype(str).str.contains('RA|DZ', regex=True), 'precipitation'] = 1
df.loc[df['weather_2'].astype(str).str.contains('RA|DZ', regex=True), 'precipitation'] = 1
df.loc[df['weather_3'].astype(str).str.contains('RA|DZ', regex=True), 'precipitation'] = 1
df.head(50)
```

```python
df.tail(50)
```

```python
df['obscuration'] = 0
df.loc[df['weather_2'].astype(str).str.match('BR|FG|BCFG'), 'obscuration'] = 1
df.loc[df['weather_1'].astype(str).str.match('BR|FG|BCFG'), 'obscuration'] = 1
df.loc[df['weather_3'].astype(str).str.match('BR|FG|BCFG'), 'obscuration'] = 1
df.head(50)
```

```python
df.tail(50)
```

```python
df['thunderstorm'] = 0
df.loc[df['weather_1'].astype(str).str.match(r'(\+|-)?(TS|TSRA)'), 'thunderstorm'] = 1
df.loc[df['weather_2'].astype(str).str.match(r'(\+|-)?(TS|TSRA)'), 'thunderstorm'] = 1
df.loc[df['weather_3'].astype(str).str.match(r'(\+|-)?(TS|TSRA)'), 'thunderstorm'] = 1
df.head(50)
```

```python
df.tail(50)
```

```python
df["limited_visibility"] = 0
df.loc[df["visibility_m"].astype(float) < 5000.0, "limited_visibility"] = 1
df.head(50)
```

```python
df.tail(50)
```

```python
data['precipitation'] = 0
data.loc[data['weather_1'].astype(str).str.contains('RA|DZ', regex=True), 'precipitation'] = 1
data.loc[data['weather_2'].astype(str).str.contains('RA|DZ', regex=True), 'precipitation'] = 1
data.loc[data['weather_3'].astype(str).str.contains('RA|DZ', regex=True), 'precipitation'] = 1
data['obscuration'] = 0
data.loc[data['weather_2'].astype(str).str.match('BR|FG|BCFG'), 'obscuration'] = 1
data.loc[data['weather_1'].astype(str).str.match('BR|FG|BCFG'), 'obscuration'] = 1
data.loc[data['weather_3'].astype(str).str.match('BR|FG|BCFG'), 'obscuration'] = 1
data['thunderstorm'] = 0
data.loc[data['weather_1'].astype(str).str.match(r'(\+|-)?(TS|TSRA)'), 'thunderstorm'] = 1
data.loc[data['weather_2'].astype(str).str.match(r'(\+|-)?(TS|TSRA)'), 'thunderstorm'] = 1
data.loc[data['weather_3'].astype(str).str.match(r'(\+|-)?(TS|TSRA)'), 'thunderstorm'] = 1
data["limited_visibility"] = 0
data.loc[data["visibility_m"].astype(float) < 5000.0, "limited_visibility"] = 1
data.head(50)
```

```python
data.tail(50)
```
