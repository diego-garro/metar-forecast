```python
import pandas as pd
STATION = "mrpv"
```

## Velocidad y ráfagas de viento

Revisando si hay datos anómalos que se peuedan corregir

```python
wind_speed = data[["Date", "Wind_speed", "Wind_gust"]]
print(wind_speed.shape)
print(wind_speed.head())
```

### Velocidad del viento

```python
print(wind_speed["Wind_speed"].max())
print(wind_speed["Wind_speed"].min())
```

```python
anomalias = wind_speed.query("Wind_speed > 30.0")
print(anomalias.shape)
print(anomalias.head(anomalias.shape[0]))
```

```python
print(wind_speed["Wind_gust"].max())
print(wind_speed["Wind_gust"].min())
```

```python
anomalias = wind_speed.query("Wind_gust > 45.0")
print(anomalias.shape)
print(anomalias.head(anomalias.shape[0]))
```
